```python
import time
import pandas as pd
from ftx_client import FtxClient
import datetime
import matplotlib.pyplot as plt
import requests
import numpy as np
from statsmodels.tsa.stattools import adfuller
```

```python
def get_ftx_dfs():
    ftx_data = requests.get(url="https://ftx.com/api/markets").json()
    df = pd.DataFrame(ftx_data['result'])
    df = df.sort_values(by='change1h', ascending=False)
    df = df.drop(columns=['underlying', 'type', 'minProvideSize', 'sizeIncrement', 'last', 'priceIncrement', 'changeBod', 'postOnly', 'restricted', 'ask', 'bid', 'baseCurrency', 'quoteCurrency', 'highLeverageFeeExempt', 'enabled', 'largeOrderThreshold', 'change24h', 'quoteVolume24h', 'priceHigh24h', 'priceLow24h', 'tokenizedEquity'])
    perps = df.copy()[df['name'].str.contains('-PERP')]

    return perps
```

```python
per = get_ftx_dfs()
print(per.to_string())
```
